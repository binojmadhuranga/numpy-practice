```python
import numpy as np
import time

size = 1_000_000
np_array = np.arange(size)
py_list = list(range(size))
```

```python
start = time.time()
np_array_result = np_array + 5
end = time.time()
print("NumPy array time:", end - start)
```

```python
start = time.time()
py_list_result = [x + 5 for x in py_list]
end = time.time()
print("Python list time:", end - start)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40])

# Addition
print(arr + 5)      # add 5 to every element

# Subtraction
print(arr - 5)      # subtract 5 from every element

# Multiplication
print(arr * 2)      # multiply every element by 2

# Division
print(arr / 10)     # divide every element by 10
```

```python
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print(a + b)     # array addition
print(a - b)     # array subtraction
print(a * b)     # element-wise multiplication
print(a / b)     # element-wise division
```

```python
arr2 = np.array([
    [1, 2],
    [3, 4]
])

print(arr2 + 10)    # add to all elements
print(arr2 * 3)     # multiply all elements
```

```python
import numpy as np

arr = np.array([1, 2, 3])
copy_arr = arr.copy()

copy_arr[0] = 100

print("Original:", arr)     # [1 2 3]
print("Copy:", copy_arr)    # [100 2 3]
```

```python
arr = np.array([1, 2, 3])
view_arr = arr.view()

view_arr[0] = 100

print("Original:", arr)     # [100 2 3]
print("View:", view_arr)    # [100 2 3]
```

```python
import numpy as np
import time

size = 1_000_000
np_array = np.arange(size)
py_list = list(range(size))
```

```python
start = time.time()
np_array_result = np_array + 5
end = time.time()
print("NumPy array time:", end - start)
```

```python
start = time.time()
py_list_result = [x + 5 for x in py_list]
end = time.time()
print("Python list time:", end - start)
```
